### Signal Simulation formula from Tort et. al

```python
N = 8000
fs = 2000
timepoints = np.arange(N) / fs
x1 = generate_coupled_signal(10, 250, 1, 1/6, 0.01, timepoints, 1)
x2 = generate_coupled_signal(24, 350, 0.7, 1/9, 0.02, timepoints, 1)

x = x1 + x2
```

```python
cut = 4
cut = int(cut * fs)
plt.figure(figsize=(8, 4))
plt.plot(timepoints[:cut], x[:cut], color='black')
plt.grid(linestyle='--')
plt.xlabel("Время, с")
plt.ylabel("Амлитуда, мкВ")

sns.set(context='notebook', style='ticks', palette='bright', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```
